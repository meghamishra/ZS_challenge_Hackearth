```python
def smape(A, F):
    return 100/len(A) * np.sum(2 * np.abs(F - A) / (np.abs(A) + np.abs(F)))
```

```python
y=dfr['Sales']
dfr=dfr.drop(["Sales"],axis=1)
```

```python
dfr=dfr.astype(int)
dfrt=dfrt.astype(int)
```

```python
import sklearn.cross_validation
import sklearn.grid_search
import sklearn.metrics
import sklearn.neighbors
from sklearn.ensemble import RandomForestRegressor,  GradientBoostingRegressor
import sklearn.decomposition
import sklearn
import pandas as pd

X_train, X_test, Y_train, Y_test = sklearn.cross_validation.train_test_split(dfr,y,test_size=0.33,random_state=42)

k = np.arange(100)+1
parameters = {'n_estimators': [8,9,10,11,12,13,14,15,20,30,40,50,60,70,80,90,100,150,200],
             'bootstrap':[True,False]}
knn = RandomForestRegressor(n_jobs=-1)
clf = sklearn.grid_search.GridSearchCV(knn,parameters,cv=5,n_jobs=-1,verbose=1)

clf.fit(X_train,Y_train)
yt=clf.predict(X_test)
print(smape(yt,Y_test))

print(clf.best_params_)
```
